```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Ganti 'california_housing_train.csv' dengan nama file CSV yang sesuai dan pastikan berada di lokasi yang benar.
file_path = '/content/sample_data/california_housing_test.csv'

# Membaca dataset ke dalam DataFrame
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama dari DataFrame
df.head()
```

```python
df.info()
```

```python
# Nama fitur-fitur yang ingin Anda hitung jumlah nilai uniknya
features = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']

# Menghitung jumlah nilai unik untuk masing-masing fitur
unique_value_counts = df[features].nunique()

# Menampilkan hasil
print(unique_value_counts)
```

```python
summary = df.describe()
print(summary)
```

```python
# Membaca daftar fitur yang ingin divisualisasikan
fitur_visual = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']

# Memvisualisasikan distribusi untuk setiap fitur
for feature in fitur_visual:
    plt.figure(figsize=(8, 6))  # Ukuran grafik
    sns.histplot(df[feature], kde=True, color='skyblue')  # Menggunakan histogram dan KDE
    plt.title(f'Distribusi dari {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frekuensi')
    plt.show()
```

```python
# Daftar fitur yang ingin divisualisasikan
fitur_visual = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']

# Membuat visualisasi histogram interaktif dengan kontrol bin width
for feature in fitur_visual:
    fig = px.histogram(df, x=feature, title=f'Distribusi dari {feature}')
    fig.update_traces(marker_color='skyblue')
    fig.update_xaxes(type="category")

    # Menambahkan kontrol bin width yang interaktif
    fig.update_xaxes(categoryorder='total ascending')
    fig.update_xaxes(title_text='Bin Width')

    fig.show()
```

```python
# Membaca daftar fitur yang ingin divisualisasikan
fitur_visual = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']

# Mengatur subplot 3x3
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.subplots_adjust(wspace=0.4, hspace=0.4)

# Memvisualisasikan boxplot untuk setiap fitur dengan orientasi horizontal dan warna ungu
for i, feature in enumerate(fitur_visual):
    row = i // 3
    col = i % 3
    sns.boxplot(data=df, x=feature, color='purple', orient='h', ax=axes[row, col])
    axes[row, col].set_title(f'Boxplot of {feature}')
    axes[row, col].set_xlabel(feature)
    axes[row, col].set_ylabel('')

# Mengatur tata letak subplot
plt.tight_layout()
plt.show()
```

```python
# Membuat scatter plot untuk setiap pasang fitur
sns.set(style="ticks")
sns.pairplot(df, kind="scatter", diag_kind="kde", palette="husl")
plt.show()
```

```python
# Membuat heatmap untuk setiap fitur
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Korelasi Fitur', size=15)
plt.show()
```

```python
# Mengatur tata letak secara rapi
plt.figure(figsize=(12, 8))

# Membuat violin chart untuk setiap fitur
for i, column in enumerate(df.columns):
    if df[column].dtype == 'float64':
        plt.subplot(3, 3, i+1)
        sns.violinplot(y=df[column], color='skyblue')
        plt.title(f'Violin Chart {column}', size=12)
        plt.ylabel(column, size=10)

# Menampilkan plot
plt.tight_layout()
plt.show()
```
